```python
#19bec029 19bec035 - 8086 Assembler using Python

import tkinter as tk

# (Enter in format - INSTRUCTION ARGUMENT_1,ARGUMENT2)
instruction = "" #input instruction string

#Setting Up GUI
frame = tk.Tk()
frame.title("8086 Assembly Code")
frame.geometry('1000x500')

def printInput():
    inp = inputtxt.get(1.0, "end-1c")
    lbl.config(text = "Input Code: \n"+inp)
    global instruction
    instruction =inp
# TextBox Creations
inputtxt = tk.Text(frame,height = 20,width = 40)
inputtxt.pack()
# Button Creation
printButton = tk.Button(frame,text = "Run",command = printInput)
printButton.pack()
# Label Creation
lbl = tk.Label(frame, text = "CODE- ")
lbl.pack()
frame.mainloop()

data = instruction.splitlines()

#Lookup Tables
Opcode = {'MOV' : 1004,'ADD' : 1008, 'INC' : 1012, 'SUB' : 1016, 'DEC' : 1020, 'MUL' : 1024, 'DIV':1028,'XCHG':1032,'XOR' : 1036, 'CLI' : 1040, 'SAL' : 1044,'NOT' : 1048, 'OR' : 1052 ,'AND' : 1056, 'CLC': 1060 }
reg = {'AL' : 0, 'AH' : 0, 'BL' : 0, 'BH' : 0, 'CL' : 0, 'CH' : 0, 'DL' : 0, 'DH' : 0,'AX':0,'BX':0,'CX':0,'DX':0}
flags = { 'CF' : 0, 'PF' : 0, 'AF' : 0, 'ZF' : 0, 'SF' : 0,'OF' : 0,'TF' : 0, 'IF' : 0, 'DF' : 0}

def print_all():  #To Print all registers and flag
    global reg
    print("Registers:- ")
    for i in reg:
        print(i,hex(reg[i]),end=", ")
    print("\nFlags:- ",flags)
    
#Instruction Set functions
def mov(reg1,reg2):
    global reg
    if reg2 in reg.keys():
        reg[reg1] = reg[reg2]
    else:
        reg[reg1] = int(reg2) 
    
def add(reg1,reg2):
    global reg
    if reg2 in reg.keys():
        reg[reg1] = reg[reg1]+reg[reg2]

    else:
        reg[reg1] = reg[reg1]+int(reg2)
        
def inc(reg1):
    reg[reg1] = reg[reg1]+1

def sub(reg1,reg2):
    global reg
    if reg2 in reg.keys():
        reg[reg1] = reg[reg1]-reg[reg2]
    else:
        reg[reg1] = reg[reg1]-int(reg2)
    
def dec(reg1):
    reg[reg1] = reg[reg1]-1
    
def mul(reg1):
    reg['AX'] = reg['AX']*reg[reg1]
    
def div(reg1):
    reg['AL'] = reg['AX']//reg[reg1]
    reg['AH'] = reg['AX']%reg[reg1]
    
def xchg(reg1,reg2):
    reg[reg1],reg[reg2] = reg[reg2],reg[reg1]
    
def and_fun(reg1,reg2):
    reg[reg1] = reg[reg1]&eg[reg2]
    
def or_fun(reg1,reg2):
    reg[reg1] = reg[reg1]|reg[reg2]

def xor_fun(reg1,reg2):
    reg[reg1] = reg[reg1]^reg[reg2]

def cli():
    flags['IF']=0

def not_fun(reg1):
    reg[reg1] = ~reg[reg1]
    
def sal(reg1,reg2):
    reg[reg1] = reg[reg1]<<reg2
        
def clc():
    flags['CF'] = 0
    
    
#Main Body
for s in data:
    global reg
    s = s.upper()
    x = s.split()
    inst = x[0];
    if len(x)>1:  #If instruction consists of registers
        temp = x[1].split(',')
        reg1= temp[0]
        if reg1 not in reg.keys():
            if len(temp)==1:
                print("\n >>",inst,reg1)
            #Checking for valid registers via lookup table
            else:    
                print("\n >>",inst,reg1,temp[1])
            print("ERROR: ____________INVALID REGISTER______________")
            continue
        if len(temp)==1:
            print("\n >>",inst,reg1)
        if len(temp)==2:
            reg2=temp[1]
            if reg2.isalpha():
                if reg2 not in reg.keys():
                    print("\n >>",inst,reg1,reg2)
                    print("ERROR: ____________INVALID REGISTER______________")
                    continue
            print("\n >>",inst,reg1,reg2)
        elif len(temp)>2:
            print("Syntax Error")
    else:
        print("\n >>",inst)
   
    if inst not in Opcode.keys():
        #Checking for valid intruction set via lookup table
        print("ERROR: ______________INSTRUCTION NOT FOUND!______________")
        continue
    #Matching the instruction through lookup table
    elif Opcode[inst]==1004:
        mov(reg1,reg2)
    elif Opcode[inst]==1008:
        add(reg1,reg2)
    elif Opcode[inst]==1012:
        inc(reg1)
    elif Opcode[inst]==1016:
        sub(reg1,reg2)
    elif Opcode[inst]==1020:
        dec(reg1)
    elif Opcode[inst]==1024:
        mul(reg1)
    elif Opcode[inst]==1028:
        div(reg1)
    elif Opcode[inst]==1032:
        xchg(reg1,reg2)
    elif Opcode[inst]==1036:
        xor_fun(reg1,reg2)
    elif Opcode[inst]==1040:
        cli()
    elif Opcode[inst]==1044:
        sal(reg1,reg2)
    elif Opcode[inst]==1048:
        not_fun(reg1)
    elif Opcode[inst]==1052:
        or_fun(reg1,reg2)
    elif Opcode[inst]==1056:
        and_fun(reg1,reg2)
    elif Opcode[inst]==1060:
        clc()
    else:
        print("Error Occured!")
        
    print_all();
```
